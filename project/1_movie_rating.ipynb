```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./datefile/movies.csv",parse_dates=["Release_Date"])
df
```

```python
df["Year"] = df["Release_Date"].dt.year
df
```

```python
avg_genre = df.groupby(["Genre"])["Rating"].mean().sort_values(ascending=False)
avg_genre
```

1. Average Rating by Genre

```python
plt.figure(figsize=(8,5))
# sns.barplot(x=avg_genre.index,y=avg_genre.values,palette='viridis')
plt.bar(avg_genre.index,avg_genre.values)
plt.title("Average Rating by Genre")
plt.ylabel("Average Rating")
plt.xlabel("Genre")
plt.grid(linestyle=":",color="gray",linewidth=0.8)
plt.xticks(rotation=45)
plt.show()
```

2. Average Rating by Year

```python
avg_rating_year = df.groupby(["Year"])["Rating"].mean()
avg_rating_year
```

```python
plt.plot(avg_rating_year.index,avg_rating_year.values,marker="o",color="b")
plt.title("Average Rating by Year")
plt.ylabel("Rating")
plt.xlabel("Year")
plt.legend(["Average Year"])
plt.tight_layout()
plt.show()
```

3. Budget vs Rating Scatter Plot

```python
plt.scatter(df["Budget"],df["Rating"])
plt.title("Budget vs Rating")
plt.ylabel("Rating")
plt.xlabel("Budget (usd)")
plt.xscale("log")
plt.legend(["Genre"])  
plt.grid(linestyle=":")
plt.show()
```
