```python
import matplotlib.pyplot as plt
import datetime
```

- line chart

```python
Student = ['A','B','C','D']
Mark = [90,98,76,43]

plt.plot(Student,Mark,marker='o',color='red')
plt.xlabel("Student")
plt.ylabel("Mark")
plt.title("Student Mark")
plt.grid()
plt.legend(["Mark"])
plt.show()
```

```python
dates = [
    datetime.date(2025, 9, 20),
    datetime.date(2025, 9, 21),
    datetime.date(2025, 9, 22),
    datetime.date(2025, 9, 23),
    datetime.date(2025, 9, 24)
]

values = [100, 120, 115, 130, 125]
plt.plot(dates,values,linestyle='-',color='blue',marker='o')
plt.grid(linestyle=':')
plt.xticks(rotation=45)
plt.title("line charts")
plt.tight_layout()
plt.legend(["Date"])
plt.xlim(datetime.date(2025, 9, 20),datetime.date(2025, 9, 23))
plt.ylim(110,130)
plt.show()
```

- bar chart 

```python
# vertical bar chart
plt.bar(Student,Mark,color='b')
plt.xlabel("Student")
plt.ylabel("Student Mark")
plt.legend("Student Mark")
plt.grid(linestyle=':',linewidth='.5')
plt.show()
```

```python
# horizontal bar chart
plt.barh(Student,Mark,color='b')
plt.xlabel("Student")
plt.ylabel("Student Mark")
plt.legend("Student Mark")
plt.tight_layout()
plt.grid(linestyle=':',linewidth='.5')
plt.show()
```

- pie chart 

```python
labels = ['Food', 'Rent', 'Utilities', 'Entertainment']
values = [30, 40, 20, 10]

plt.pie(values,labels=labels,autopct='%1.1f%%')
plt.title("Monthly Budget")
plt.show()
```

- histogram chart

```python
ages = [12, 14, 16, 18, 19, 21, 22, 22, 23, 25, 26, 27, 28, 29, 30, 32, 35, 36, 38, 40]
plt.hist(ages,bins=5,edgecolor='w',color='b')
plt.tight_layout()
plt.show()
```

- scatter chart

```python
age = [18, 22, 35, 30, 25]
minutes_watched = [120, 90, 250, 300, 180]
 
plt.scatter(age,minutes_watched)
plt.title("Netflix Usage by Age")
plt.xlabel("Age of viewers")
plt.ylabel("Watch time of viewers")
plt.grid(linestyle=":")
plt.tight_layout()
plt.legend(["Viewers Age"])

plt.show()
```

```python
import numpy as np

plt.scatter(np.random.randint(10,size=5),np.random.randint(100,size=5),color="b",label='class a')
plt.scatter(np.random.randint(10,size=5),np.random.randint(100,size=5),color="r",label='class b')
plt.legend()
plt.grid(linestyle=':')
plt.xlabel("Student")
plt.ylabel("Student mark")
plt.show()
```

- subplot

```python
x = [1,2,3,4]
y = [10,20,5,30]

plt.subplot(1,2,1)
plt.plot(x,y)

plt.subplot(1,2,2)
plt.bar(x,y)

plt.tight_layout()
plt.show()
```

```python
fig,ax = plt.subplots(1,2,figsize=(10,5))

ax[0].plot(x,y,color="r",marker="o")
ax[0].set_title("Line chart")


ax[1].bar(x,y,color="b")
ax[1].set_title("bar chart")

fig.suptitle("comparison")

plt.show()
```
