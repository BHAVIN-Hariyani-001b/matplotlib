```python
import matplotlib.pyplot as plt
```

```python
# plt.subplot(rows, ncolumn, index)

x = [1,2,3,4]
y = [10,20,5,30]

plt.figure(figsize=(10,5)) # it is use to size of set to canvas(charts)

plt.subplot(1,2,1) # it is use to create plot of charts
plt.plot(x,y) # subplot create chart's
plt.title('line charts')

plt.subplot(1,2,2)
plt.bar(x,y)

plt.tight_layout() # it is use to set the space between the charts
plt.title('bar charts')
plt.show()
```

```python
# object oriented interface

x = [1,2,3,4]
y = [10,20,5,30]

# fig,ax = plt.subplot(nrows,ncols,figsize=(width,height))

fig,ax = plt.subplots(1,2,figsize=(10,5)) # it is use to size of set to canvas(charts)

ax[0].plot(x,y,color='r',marker='o') # subplot create chart's and set color and marker of line
ax[0].set_title("Line Plot")

ax[1].bar(x,y,color='g')
ax[1].set_title("Bar charts") # set title of charts

fig.suptitle("Comparison of Line and Bar Charts") # set title of figure

plt.tight_layout() # it is use to set the space between the charts
plt.show()
```

```python
plt.plot([1,2,3,4],[10,20,5,30],marker='o',color='r') # set marker and color of line
plt.title("line charts")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend(["line charts"])

plt.savefig("./charts/line_chart.png",dpi=300,bbox_inches='tight') # it is use to save the charts
# dpi = dots per inch and it is use to set the resolution of the image
# bbox_inches = 'tight' it is use to remove the extra white space and it is use to fit the image

plt.show()
```
