```python
import matplotlib.pyplot as plt
```

```python
# plt.bar(X,hight,color='',width='',label=')
product = ['A','B','C','D']
sales = [1000,1500,800,1200]

# plt.bar(product,sales,color='orange',label="Sales 2025",width=.6) # vertical 
plt.barh(product,sales,color='orange',label="Sales 2025") # horizontal
plt.xlabel("Product")
plt.ylabel("Seles")
plt.title("Product selas compare")
plt.legend()
plt.show()
```

```python
# plt.pie(value,lable,color,autopec="%1.1f%%")
regions = ['North','South','East','West']
revenue = [3000,2000,1500,1000]

plt.pie(revenue,labels=regions,autopct='%1.1f%%',colors=['gold','skyblue','lightgreen','coral'])
plt.title("Revenue comtribution by regionos")
plt.show()
```

```python
# histogram chart
# plt.hist(data,bins=number_of_bins,color='',edgcolor='')
scores = [45,67,89,56,78,88,92,60,74,81,59,66,90,85,61,77]
plt.hist(scores,bins=5,color='purple',edgecolor='black')
plt.xlabel('score range')
plt.ylabel('number of student')
plt.title('score distribution of student'.capitalize())
plt.show()
```
