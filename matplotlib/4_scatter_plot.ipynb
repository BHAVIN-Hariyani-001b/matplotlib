```python
import matplotlib.pyplot as plt
```

```python
# plt.scatter(x,y,color='',marker='',label='')
hours_student = [1,2,3,4,5,6,7,8]
exam_score = [50,55,60,65,70,75,80,85]
plt.scatter(hours_student,exam_score,color='red',marker='o',label='Student Date')
plt.xlabel("Hours study")
plt.ylabel("Exam score")
plt.title("Relation between study time and exam score")
plt.grid()
plt.legend()
plt.show()
```
